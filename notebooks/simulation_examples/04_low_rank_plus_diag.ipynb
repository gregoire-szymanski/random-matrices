# Example 04: Low-Rank Plus Diagonal Covariance

This run uses Cov = B Sigma B^T + D and visualizes realized-vs-target spectra.

```python
from pathlib import Path
import numpy as np
import matplotlib.pyplot as plt

from mpdiff.experiments.run_simulation import run_simulation

CONFIG_PATH = Path('../../configs/simulation_examples/04_low_rank_plus_diag.yaml')
summary = run_simulation(CONFIG_PATH)
summary
```

```python
output_dir = Path(summary['output_dir'])
times = np.load(output_dir / 'times.npy')
paths = np.load(output_dir / 'paths.npy')
realized_eigs = np.load(output_dir / 'realized_eigenvalues.npy')
target_eigs = np.load(output_dir / 'target_population_eigenvalues.npy')

plt.figure(figsize=(9,4))
for k in range(min(4, paths.shape[1])):
    plt.plot(times, paths[:, k], label=f'X[{k}]')
plt.title('Sample Path Coordinates')
plt.xlabel('time')
plt.legend(frameon=False, ncol=2)
plt.grid(alpha=0.2)
plt.show()

plt.figure(figsize=(9,4))
plt.hist(realized_eigs, bins=50, density=True, alpha=0.5, label='realized')
plt.hist(target_eigs, bins=50, density=True, alpha=0.4, label='target population')
plt.title('Eigenvalue Histogram: Realized vs Target')
plt.xlabel('eigenvalue')
plt.legend(frameon=False)
plt.grid(alpha=0.2)
plt.show()
```
