# Example 06: Piecewise Scalar-Times-Base Covariance

This run uses Cov_j = c_j * M_j with independent base draws across segments.

```python
from pathlib import Path
import sys

PROJECT_ROOT = next(
    (
        p
        for p in (Path.cwd().resolve(), *Path.cwd().resolve().parents)
        if (p / 'pyproject.toml').exists() and (p / 'src' / 'mpdiff').exists()
    ),
    None,
)
if PROJECT_ROOT is None:
    raise RuntimeError('Could not locate project root (expected pyproject.toml and src/mpdiff).')
SRC_PATH = PROJECT_ROOT / 'src'
if str(SRC_PATH) not in sys.path:
    sys.path.insert(0, str(SRC_PATH))

import numpy as np
import matplotlib.pyplot as plt

from mpdiff.experiments.run_simulation import run_simulation

CONFIG_PATH = PROJECT_ROOT / 'configs/simulation_examples/06_piecewise_scaled_base.yaml'
summary = run_simulation(CONFIG_PATH)
summary
```

```python
output_dir = Path(summary['output_dir'])
times = np.load(output_dir / 'times.npy')
paths = np.load(output_dir / 'paths.npy')
realized_eigs = np.load(output_dir / 'realized_eigenvalues.npy')
target_eigs = np.load(output_dir / 'target_population_eigenvalues.npy')

plt.figure(figsize=(9,4))
for k in range(min(4, paths.shape[1])):
    plt.plot(times, paths[:, k], label=f'X[{k}]')
plt.title('Sample Path Coordinates')
plt.xlabel('time')
plt.legend(frameon=False, ncol=2)
plt.grid(alpha=0.2)
plt.show()

plt.figure(figsize=(9,4))
plt.hist(realized_eigs, bins=55, density=True, alpha=0.5, label='realized')
plt.hist(target_eigs, bins=55, density=True, alpha=0.4, label='target population')
plt.title('Eigenvalue Histogram: Realized vs Target')
plt.xlabel('eigenvalue')
plt.legend(frameon=False)
plt.grid(alpha=0.2)
plt.show()
```
